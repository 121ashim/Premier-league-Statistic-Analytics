```python
#  WHO CARRIES THIER TEAM
# Make a copy to avoid modifying the original df
df_copy = df.copy()


club_stats = df_copy.groupby('Club')[['Goals', 'Assists']].sum().reset_index()
club_stats.columns = ['Club', 'Team_Goals', 'Team_Assists']


df_copy = df_copy.merge(club_stats, on='Club')


df_copy['Goal_Contribution_Pct'] = ((df_copy['Goals'] + df_copy['Assists']) /
                                    (df_copy['Team_Goals'] + df_copy['Team_Assists'])) * 100

# filtering players with minimum  playing time
key_players = df_copy[df_copy['Minutes'] > 900].copy()
key_players = key_players.sort_values('Goal_Contribution_Pct', ascending=False).head(10)


key_players_display = key_players[['Player Name', 'Club','Position', 'Goals', 'Assists', 'Goal_Contribution_Pct']]
display(key_players_display)
```

```python
# in Bar plot
import seaborn as sns
plt.figure(figsize=(12,6))
sns.barplot(data=key_players, x='Goal_Contribution_Pct', y='Player Name', hue='Club', dodge=False, palette='viridis')
plt.title('Top 10 Most Essential Players to Their Clubs')
plt.xlabel('Goal Contribution (%)')
plt.ylabel('Player')
plt.xlim(0, 100)
plt.grid(axis='x')
plt.show()
```

```python
# most defensive team



df = pd.read_csv("/content/drive/MyDrive/football/epl_player_stats_24_25.csv")


df['DefensiveMetric'] = (
    (-df['Goals Conceded']) +
    df['Clean Sheets'] +
    df['Tackles'] +
    df['Interceptions'] +
    df['Clearances'] +
    df['Blocks'] +
    df['Possession Won'] +
    0.5 * df['Saves'] +
    df['Goals Prevented']
)


team_defense = df.groupby("Club")['DefensiveMetric'].sum().sort_values(ascending=False)


best_team = team_defense.head(10)
best_score = team_defense.max()

print("Most Defensive Team:")
print(f"{best_team}  â†’  Defensive Score: {best_score:.2f}")


```

```python
plt.figure(figsize=(12,6))
team_defense.head(10).plot(kind='bar')
plt.title("Top 10 Most Defensive Teams (Custom Metric)")
plt.xlabel("Team")
plt.ylabel("Defensive Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/football/epl_player_stats_24_25.csv")

# Filter only midfielders (case-insensitive)
mid_df = df[df["Position"].str.contains("MID", case=False, na=False)]

# Extract metrics
x = mid_df["Passes"]
y = mid_df["Progressive Carries"]

plt.figure(figsize=(12, 7))

# Scatter plot
plt.scatter(x, y)

plt.title("\ Passes vs Progressive Carries", fontsize=14)
plt.xlabel("Total Passes", fontsize=12)
plt.ylabel("Progressive Carries", fontsize=12)

# Annotate each midfielder
for i, txt in enumerate(mid_df["Player Name"]):
    plt.annotate(txt, (x.iloc[i], y.iloc[i]), fontsize=9, alpha=0.7)


plt.tight_layout()
plt.show()

# top right= ball playing CM
# top left = box-box CM
# bottom right=  deep line playmaker
# bottom left = destroyers
```

```python
# FILTER â†’ Attackers
atk = df[df["Position"] == "FWD"]

x = atk["Passes"]
y = atk["Progressive Carries"]

plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.title("Attackers: Passes vs Progressive Carries")
plt.xlabel("Total Passes")
plt.ylabel("Progressive Carries")

# Player labels
for i, txt in enumerate(atk["Player Name"]):
    plt.annotate(txt, (x.iloc[i], y.iloc[i]), fontsize=9, alpha=0.7)

plt.tight_layout()
plt.show()

#op-right â†’ ball-carrying creators
#Bottom-right â†’ link-up strikers
#Top-left â†’ dribblers
#Bottom-left â†’ low-involvement finishers
```

```python
# FILTER â†’ Centre-backs
cb = df[df["Position"].isin(["DEF"])]

x = cb["Passes"]
y = cb["Progressive Carries"]

plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.title("Centre-backs: Passes vs Progressive Carries")
plt.xlabel("Total Passes")
plt.ylabel("Progressive Carries")

for i, txt in enumerate(cb["Player Name"]):
    plt.annotate(txt, (x.iloc[i], y.iloc[i]), fontsize=9, alpha=0.7)

plt.tight_layout()
plt.show()

#High passes + high carries â†’ ball-playing CB(top right)
#High passes + low carries â†’ deep distributors(Bottom right)
#Low passes + high carries â†’ aggressive stepping CB(top left)
#Low + low â†’ classic defenders(bottom left)
```
